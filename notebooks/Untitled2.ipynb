```python
from binance.client import Client
from binance.websockets import BinanceSocketManager
```

```python
client = Client("", "")
bm = BinanceSocketManager(client)
```

```python
levels=20
interval='1m'
```

```python
quote_asset = "BTC"
depth_ws_suffix = "@depth"+str(levels)
kline_ws_suffix = "@kline_"+str(interval)
```

```python
def check_quote_asset(quote_asset, symbol):
    return symbol.endswith(quote_asset);
```

```python
info = client.get_exchange_info()
linfo = [x['symbol'] for x in info['symbols'] if check_quote_asset(quote_asset,x['symbol'])]
```

```python
depth_ws_list = [y.lower()+depth_ws_suffix for y in linfo]
```

```python
def get_depth(data):
        full_depth = {}
        bids = [{'bid_price_'+str(i):float(d[0]), 'bid_quantity_'+str(i):float(d[1])} for (i, d) in enumerate(data['bids'])]
        asks = [{'ask_price_'+str(i):float(d[0]), 'ask_quantity_'+str(i):float(d[1])} for (i, d) in enumerate(data['asks'])]
        for bid, ask in zip(bids, asks):
            full_depth.update(bid)
            full_depth.update(ask)
        return full_depth
```

```python
def process_depth(msg):
        try:
            symbol = msg['stream'].replace(depth_ws_suffix, '').upper()
            depth = get_depth(msg['data'])
            print(depth)
        except Exception as e:
            print(e)
```

```python
depth_key = bm.start_multiplex_socket(
    depth_ws_list,
    process_depth
)
```

```python
bm.start()
```

```python
bm.stop_socket(depth_key)
```
